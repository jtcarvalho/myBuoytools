```python
import matplotlib.pyplot as plt
import numpy as np
from windrose import WindroseAxes
import xarray as xr
import pandas as pd
from datetime import datetime


def getVectComponents(direction):
    return -np.sin(direction), -np.cos(direction)

def replace_comma():
       text = open("data/buoy/ISPRA2440.csv", "r")
       text = ''.join([i for i in text]).replace(",", ".")
       x = open("dir.csv", "w")
       x.writelines(text)
       x.close()
       return


# Load the CSV file into a Pandas DataFrame
f='/Users/jtakeo/googleDrive/myProjects/buoys/pnboia/metocean_bacia-de-santos-bm01_qualificados_publicos.csv'

df = pd.read_csv(f, delimiter=',',skiprows=0,header=0)

#Adjust variables 
#str1=df['date_time'].astype(str)
str1=df['date_time'].values

start_date = '2020'+'-12-10 01:00:00'
end_date   = '2023'+'-08-30 02:00:00'
date_range = pd.date_range(start_date, end_date, freq='1H')

#Transforming csv time to datetime format
date_time=[]
for i in range(0,len(str1)):

    year2=str1[i][0:4]; mon=str1[i][5:7]
    day=str1[i][8:10];  hour=str1[i][11:13]
    min=str1[i][14:16];  sec='00'
    date_string = str(year2)+str(mon)+str(day)+' '+str(hour)+':'+str(min)+':'+str(sec)
    date_time.append(datetime.strptime(date_string, '%Y%m%d %H:%M:%S'))


print(len(date_range)) #time_complete
print(len(date_time))  #time_incomplete

#Comparing and filling the missing data 
hsArray= pd.to_numeric(df['swvht1'], errors='coerce').values #buoy_incomplete
data1 = pd.DataFrame(hsArray, index=date_time, columns=["hs"])
nan_array = np.full((len(date_range),), np.nan) #buoy_complete_nan
data2 = pd.DataFrame(nan_array, index=date_range, columns=["hs"])

merged_data = data1.combine_first(data2)
merged_data.loc[merged_data['hs'] < 0] = np.nan

#Correct merged index
ds = xr.Dataset.from_dataframe(merged_data)
ds = ds.set_coords('index')
ds = ds.rename({'index': 'time'})

ds.to_netcdf('./hs.nc')

#Resample data
####ds_resampled = ds.resample(time='1H').mean()
####ds_resampled.to_netcdf('./hs.nc')


plt.figure(figsize=(10, 6))  # Define o tamanho da figura

# Plote o gráfico de linha
ds.plot.scatter(x='time', y='hs', color='blue', marker='o')
```
